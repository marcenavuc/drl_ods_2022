```python
import pandas as pd
import plotly.express as px
```

## Task1

```python
task1_df = pd.read_csv('task1.csv')
task1_df = task1_df[['gamma', 'mean_total_reward']]
task1_df
```

```python
px.line(task1_df, x='gamma', y='mean_total_reward')
```

## Task2

```python
task2_df = pd.read_csv('task2.csv')
task2_df = task2_df[['gamma', 'iteration_n', 'evaluation_step_n', 'mean_total_reward_default', 'mean_total_reward_mod']]
task2_df
```

```python
px.scatter(task2_df.sort_values('iteration_n').reset_index(),
           x='index',
           y=['mean_total_reward_default', 'mean_total_reward_mod'],
           labels={'index': 'experiment', 'value': 'mean_total_reward'})
```

```python
task2_df.describe()[['mean_total_reward_default', 'mean_total_reward_mod']]
```

## Task3

```python
task3_df = pd.read_csv('task3.csv')
task3_df = task3_df[
    ['gamma',
     'iteration_n',
     'evaluation_step_n',
     'mean_total_reward_default',
     'mean_total_reward_mod',
     'time_default', 
    'time_val']
]
task3_df
```

```python
px.scatter(task3_df.sort_values('iteration_n').reset_index(),
           x='index',
           y=['mean_total_reward_default', 'mean_total_reward_mod'],
           labels={'index': 'experiment', 'value': 'mean_total_reward'})
```

```python
df = task3_df.describe()[['mean_total_reward_default', 'mean_total_reward_mod', 'time_default', 'time_val']]
df.rename(columns={
    'mean_total_reward_default': 'mean_total_reward_Policy_Iteration',
    'mean_total_reward_mod':     'mean_total_reward_Value_Iteriation',
    'time_default':              'time_Policy_Iteration',
    'time_val':                  'time_Value_Iteration',
})
```
